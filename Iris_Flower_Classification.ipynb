```python
# importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# loading dataset and viewing first and last rows
data = pd.read_csv("/content/Iris.csv")
data
df = pd.DataFrame(data)
df.head()
```

```python
df.tail()
```

```python
# dropping Id column
df.drop(columns='Id', axis=1,inplace = True)
df.head()
```

```python
# visualising classes of dataset
df.groupby(['Species'])['SepalLengthCm'].count().plot.bar()
plt.xlabel('Category')
plt.ylabel('Number of')
plt.show()
```

```python
df['Species'].value_counts()
```

# Splitting the features and target (x and y respectively)

```python
x= df.drop(columns='Species',axis =1)
y= df['Species']
print(x.shape)
print(y.shape)

```

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting all data for training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=3)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Building Model

```python

from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
dtree= dtree.fit(x_train,y_train)

print('Decesion Tree Model is Created')
```

```python
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(dtree,filled=True)
plt.show()

```

```python
# checking acuuracy while training
from sklearn.model_selection import cross_val_score
cv_dt = cross_val_score(dtree, x_train, y_train, cv = 5, scoring = 'accuracy').mean()
print('For Decision Trees Classifier :', round(cv_dt,3))
```

```python
y_pred = dtree.predict(x_test)
dfa = pd.DataFrame(y_pred)
dfa.head()

```

```python
y_test.head()
```

```python
# checking final accuracy on test set
from sklearn.metrics import accuracy_score



str(round(accuracy_score(y_test, y_pred,  normalize=True, sample_weight=None) ))
```
