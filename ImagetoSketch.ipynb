# Loading the image

```python
# importing image  using cv2 library, then presenting it with help of pylot
img = cv2.imread("actor.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title("Jong-Suk Lee")
plt.imshow(img)
plt.show()
```

# Converting image  to Gray Scale
